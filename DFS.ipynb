```python
import random
def genMaze(dim,p):
    maze = []
    for x in range(dim):
        row = []
        for y in range(dim):
            if random.random() > p:
                row.append(0)
            else :
                row.append(1)
        maze.append(row)
    maze[0][0] = 0
    maze[dim-1][dim-1] = 0
    return maze
```

```python
def DFS(data):
    
    i = 0
    j = 0
    visited = []
    fringe = [[0,0]]
    print(fringe)
    while len(fringe) != 0 :
        current = fringe.pop()
        visited.append(current)
        i = current[0]
        j = current[1]
        if i == dim-1 and j == dim-1 :
            print("goal state reached")
            print("visited nodes are")
            print(visited)
            return(visited)

            
        else :
            
            if i==0 and j==0 : #expansion of start state
                if data[i+1][j] != 1 and [i+1,j] not in visited and [i+1,j] not in fringe: #expand the lower cell
                    fringe.append([i+1,j])
                    print(fringe)
                if data[i][j+1] != 1 and [i,j+1] not in visited and [i,j+1] not in fringe: #expand the right cell
                    fringe.append([i,j+1])
                    print(fringe)
            if i>=1 and j>=1 and i<=dim-2 and j<=dim-2 : #expansion for inner cells (leaving outmost cells)
                if data[i][j-1] != 1 and [i,j-1] not in visited and [i,j-1] not in fringe: #expand the left cell
                    fringe.append([i,j-1])
                    print(fringe)
                if data[i-1][j] != 1 and [i-1,j] not in visited and [i-1,j] not in fringe: #expand the upper cell
                    fringe.append([i-1,j])
                    print(fringe)
                if data[i+1][j] != 1 and [i+1,j] not in visited and [i+1,j] not in fringe: #expand the lower cell
                    fringe.append([i+1,j])
                    print(fringe)
                if data[i][j+1] != 1 and [i,j+1] not in visited and [i,j+1] not in fringe: #expand the right cell
                    fringe.append([i,j+1])
                    print(fringe)
                    
            if i==0 and j <=dim-2 and j >= 1 : #expansion of cells in first row leaving right most cell and starting state
                if data[i][j-1] != 1 and [i,j-1] not in visited and [i,j-1] not in fringe: #expand the left cell
                    fringe.append([i,j-1])
                    print(fringe)
                if data[i+1][j] != 1 and [i+1,j] not in visited and [i+1,j] not in fringe: #expand the lower cell
                    fringe.append([i+1,j])
                    print(fringe)
                if data[i][j+1] != 1 and [i,j+1] not in visited and [i,j+1] not in fringe : #expand the right cell
                    fringe.append([i,j+1])
                    print(fringe)
                    
            if i==0 and j==dim-1 : #expansion of right topmost cell
                if data[i][j-1] != 1 and [i,j-1] not in visited and [i,j-1] not in fringe: #expand the left cell
                    fringe.append([i,j-1])
                    print(fringe)
                if data[i+1][j] != 1 and [i+1,j] not in visited and [i+1,j] not in fringe: #expand the lower cell
                    fringe.append([i+1,j])
                    print(fringe)
                    
            if i<=dim-2 and i>=1 and j==0 : #expansion of cells in first column leaving bottom most cell and start state
                if data[i-1][j] != 1 and [i-1,j] not in visited and [i-1,j] not in fringe: #expand the upper cell
                    fringe.append([i-1,j])
                    print(fringe)
                if data[i+1][j] != 1 and [i+1,j] not in visited and [i+1,j] not in fringe: #expand the lower cell
                    fringe.append([i+1,j])
                    print(fringe)
                if data[i][j+1] != 1 and [i,j+1] not in visited and [i,j+1] not in fringe: #expand the right cell
                    fringe.append([i,j+1])
                    print(fringe)
                    
            if i==dim-1 and j!=0 and j<=dim-2 : #expansion of bottom most row except goal state and leftmost state
                if data[i][j-1] != 1 and [i,j-1] not in visited and [i,j-1] not in fringe: #expand the left cell
                    fringe.append([i,j-1])
                    print(fringe)
                if data[i-1][j] !=1 and [i-1,j] not in visited and [i-1,j] not in fringe: #expand the upper cell
                    fringe.append([i-1,j])
                    print(fringe)
                if data[i][j+1] != 1 and [i,j+1] not in visited and [i,j+1] not in fringe: #expand the right cell
                    fringe.append([i,j+1])
                    print(fringe)
                    
            if j==dim-1 and i>0 and i<=dim-2 : #expansion of right most column leaving top most cell and goal state
                if data[i][j-1] !=1 and [i,j-1] not in visited and [i,j-1] not in fringe: #expand the left cell
                    fringe.append([i,j-1])
                    print(fringe)
                if data[i-1][j] !=1 and [i-1,j] not in visited and [i-1,j] not in fringe: #expand the upper cell
                    fringe.append([i-1,j])
                    print(fringe)
                if data[i+1][j] != 1 and [i+1,j] not in visited and [i+1,j] not in fringe: #expand the lower cell
                    fringe.append([i+1,j])
                    print(fringe)
                    
            if i==dim and j==0 : #expansion of left most bottom cell
                if data[i-1][j] != 1 and [i-1,j] not in visited and [i-1,j] not in fringe: #expand the upper cell
                    fringe.append([i-1,j])
                    print(fringe)
                if data[i][j+1] != 1 and [i+1,j] not in visited and [i+1,j] not in fringe: #expand the right cell
                    fringe.append([i,j+1])
                    print(fringe)
                                
    if len(fringe) == 0 :
        print("solution not found")
        return visited
                    
            
    
```

```python
dim = 35
maze = genMaze(dim,0.20)
```

```python
def draw(maze):
  blocked = "⬛"
  free = "⬜"
  for row in maze:
    tiles = ""
    for tile in row:
      tiles += blocked if tile==1 else free
    print(tiles)
```

```python
def draw_soln(maze):
    blocked = "⬛"
    free = "⬜"
    final_path = "🆒"
    for row in maze:
        tiles = ""
        for tile in row:
            #tiles += blocked if tile==1 else free
            if tile == 2:
                tiles += final_path
            if tile == 1:
                tiles += blocked
            if tile == 0:
                tiles += free
            if tile == 5:
                tiles += "🏁"
            if tile == 4:
                tiles += "🔫"
        print(tiles)
```
